```python
#4.Melakukan transformasi data (label encoding, scalling, binning)
```

```python
data_employees.dtypes
```

```python
#LABEL ENCODING
from sklearn import preprocessing
kategori = ["Last Name",
            "First Name",
            "Extension",
            "Email",
            "Office Code",
            "Job Title"]

le = preprocessing.LabelEncoder()
for feature in kategori:
    if feature in data_employees.columns.values:
        data_employees[feature] = le.fit_transform(data_employees[feature])
```

```python
#5.Feature selection/extraction
#Feature Selection
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
bmsx = np.array([data_employees["Last Name"],
                data_employees["First Name"],
                data_employees["Extension"],
                data_employees['Email'],
                data_employees["Office Code"]])
bmsx.shape
```

```python
bmsx
```

```python
bmsxt=bmsx.transpose()
bmsxt.shape
```

```python
bms_std=StandardScaler().fit_transform(bmsxt)
bms_std
```

```python
pcafs=PCA(n_components=0.70,whiten=True)
bmsfs_pca=pcafs.fit_transform(bms_std)
```

```python
print('Original number of features:', bms_std.shape[1])
print('Reduced numer of features:', bmsfs_pca.shape[1])
```

```python
datafs_pca=pd.DataFrame(bmsfs_pca, columns=["PC1","PC2"])
datafs_pca
```

```python
#Feature Extraction
from sklearn import decomposition
```

```python
pcafe=decomposition.PCA(n_components=3)
bmsfe_pca=pcafe.fit_transform(bms_std)
```

```python
print('Original number of features:', bms_std.shape[1])
print('Reduced numer of features:', bmsfe_pca.shape[1])
```

```python
datafe_pca=pd.DataFrame(bmsfe_pca, columns=["PC1","PC2", "PC3"])
datafe_pca
```
